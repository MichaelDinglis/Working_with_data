# Plotting with matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

* After importing the libraries that we are going to use, we create a numpy array with the numbers that we want to plot.

* We use <b>list(range())</b> to create a list (array) with the range of the numbers that we give in the ( )

```python
xAxis=np.array(list(range(1,10)))
yAxis=np.array(list(range(11,20)))
xAxis
yAxis
```

* Using .plot(x,y) ,  the dataFrame gets plotted

```python
plt.plot(xAxis,yAxis)
```

* Let's see how the graph comes out if the numbers are different:

```python
xAxis=np.array([1,50,30,1,2,3])
yAxis=np.array([20,3,41,3,22,75])
```

* Let's give a title, and labels for the x and y axes, and then plot the lines:
* You can also choose what is displayed on the graph on the axes using <b>.xticks() </b>and<b> .yticks()</b>
* After writting .plot(x,y), you can add a , and change the color of the plot.

```python
plt.title("Messed Up Triangle")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(xAxis,yAxis, color='green')
plt.xticks((10,20,30,40,50,60),('c','i','r','c','l','e'))
```

* You can also create more kinds of graphs, such as

```python
plt.pie(xAxis)
```

```python
plt.hist(yAxis)
```
